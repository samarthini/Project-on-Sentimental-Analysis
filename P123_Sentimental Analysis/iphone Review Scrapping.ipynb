```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
def get_soup(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.content,'html.parser')
    return soup
```

```python
reviews = []
body=[]
def get_review(soup): 
    item = soup.find_all('div',{'data-hook':'review'})
    for x in item:
        body = x.find('span',{'data-hook':'review-body'}).text.strip()
        reviews.append(body)
```

```python
for x in range(1,200):
    soup = get_soup(f'https://www.amazon.in/Apple-iPhone-4S-White-8GB/product-reviews/B00GFGKC92/ref=cm_cr_arp_d_paging_btm_next_2?pageNumber={x}')
    get_review(soup)
    
```

```python
reviews = pd.DataFrame(reviews,columns=['reviews'])
reviews
```

```python
reviews.to_csv('AmazonReviews.csv')
```
